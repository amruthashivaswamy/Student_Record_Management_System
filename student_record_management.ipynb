```python
def add_student():

    print("========== ADD STUDENT ==========\n")

    name = input("Enter Student Name: ")
    age = int(input("Enter Age: "))
    city = input("Enter City: ")
    course = input("Enter Course: ")

    Python_Marks = int(input("Enter Python Marks: "))
    SQL_Marks = int(input("Enter SQL Marks: "))
    PowerBI_Marks = int(input("Enter Power BI Marks: "))

    student = {
        "Name": name,
        "Age": age,
        "City": city,
        "Course": course,
        "Python_Marks": Python_Marks,
        "SQL_Marks": SQL_Marks,
        "PowerBI_Marks": PowerBI_Marks
    }

    students.append(student)

    print("Student Details")
    print("-"*40)
    print("Name      :", name)
    print("Age       :", age)
    print("City      :", city)
    print("Course    :", course)
    print("Python    :", Python_Marks)
    print("SQL       :", SQL_Marks)
    print("Power BI  :", PowerBI_Marks)
    print("-"*40)

    print("\n✅ Student Added Successfully!\n")
```

```python
add_student()
```

```python
def view_students():

    if len(students) == 0:
        print("No student records found.")
        return

    print("\n========== STUDENT DETAILS ==========\n")

    for student in students:
        print(f"Name            : {student['Name']}")
        print(f"Age             : {student['Age']}")
        print(f"City            : {student['City']}")
        print(f"Course          : {student['Course']}")
        print(f"Python Marks    : {student['Python_Marks']}")
        print(f"SQL Marks       : {student['SQL_Marks']}")
        print(f"Power BI Marks  : {student['PowerBI_Marks']}")
        print("-" * 40)
```

```python
view_students()
```

```python
def update_city():

    name = input("Enter the student name to update city: ")

    for student in students:

        if student["Name"].lower() == name.lower():

            new_city = input("Enter the new city: ")

            student["City"] = new_city
            print(f"Enter the student name to update city: {name}")
            print(f"Enter the new city: {new_city}")
            print("\n✅ City updated successfully!")
            return

    print("\n❌ Student not found!")
```

```python
update_city()
```

```python
def show_statistics():

    if len(students) == 0:
        print("No student records found.")
        return

    print("\n========== STUDENT STATISTICS ==========\n")

    print("Total Number of Students :", len(students))

    cities = []

    for student in students:
        if student["City"] not in cities:
            cities.append(student["City"])

    print("Cities Available :", ", ".join(cities))

    placement_count = 0

    print("\nStudent-wise Statistics")
    print("-" * 50)

    for student in students:

        marks = [
            student["Python_Marks"],
            student["SQL_Marks"],
            student["PowerBI_Marks"]
        ]

        average = sum(marks) / len(marks)
        highest = max(marks)

        print(f"Name            : {student['Name']}")
        print(f"Python Marks    : {student['Python_Marks']}")
        print(f"SQL Marks       : {student['SQL_Marks']}")
        print(f"Power BI Marks  : {student['PowerBI_Marks']}")
        print(f"Average Marks   : {average:.2f}")
        print(f"Highest Marks   : {highest}")

        if average >= 40:
            print("Placement       : Eligible")
            placement_count += 1
        else:
            print("Placement       : Not Eligible")

        print("-" * 50)

    print("Students Eligible for Placement :", placement_count)
```

```python
show_statistics()
```

```python
while True:

    print("\n" + "="*50)
    print("      STUDENT RECORD MANAGEMENT SYSTEM")
    print("="*50)
    print("1. Add Student")
    print("2. View Students")
    print("3. Update City")
    print("4. Show Statistics")
    print("5. Exit")

    choice = input("\nEnter your choice (1-5): ")

    if choice == "1":
        add_student()

    elif choice == "2":
        view_students()

    elif choice == "3":
        update_city()

    elif choice == "4":
        show_statistics()

    elif choice == "5":
        print("\nThank you for using Student Record Management System!")
        break

    else:
        print("\n❌ Invalid Choice! Please enter a number between 1 and 5.")
```
